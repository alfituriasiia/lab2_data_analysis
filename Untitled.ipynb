```python
import numpy as np
import rasterio
from rasterio.merge import merge
from rasterio.plot import show
from rasterio.warp import calculate_default_transform, reproject, Resampling
from pathlib import Path
import os
```

```python
# Папка з твоїми файлами
data_folder = r"D:\зпад\data7\S2A_MSIL2A_20190821T085601_N0213_R007_T36UUA_20190821T115206.SAFE\GRANULE"
granule_dirs = os.listdir(data_folder)
granule_path = os.path.join(data_folder, granule_dirs[0], "IMG_DATA", "R60m")

```

```python
# Знаходимо потрібні канали
files = os.listdir(granule_path)
b2 = [f for f in files if 'B02' in f][0]
b3 = [f for f in files if 'B03' in f][0]
b4 = [f for f in files if 'B04' in f][0]
b8a = [f for f in files if 'B8A' in f][0]

# Повні шляхи до файлів
b2_path = os.path.join(granule_path, b2)
b3_path = os.path.join(granule_path, b3)
b4_path = os.path.join(granule_path, b4)
b8a_path = os.path.join(granule_path, b8a)
```

```python
# Зчитуємо канали
with rasterio.open(b2_path) as src:
    b2_data = src.read(1)
    profile = src.profile

with rasterio.open(b3_path) as src:
    b3_data = src.read(1)

with rasterio.open(b4_path) as src:
    b4_data = src.read(1)

with rasterio.open(b8a_path) as src:
    b8a_data = src.read(1)
```

```python
# Записуємо об'єднаний файл
output_path = r"D:\зпад\data7\sentinel_merge.tif"
profile.update(count=4)

with rasterio.open(output_path, 'w', **profile) as dst:
    dst.write(b2_data, 1)
    dst.write(b3_data, 2)
    dst.write(b4_data, 3)
    dst.write(b8a_data, 4)

print("Об'єднання каналів завершено")
```

```python
# Показуємо об'єднаний файл (перед репроекцією)
with rasterio.open(output_path) as src:
    img = src.read([1,2,3])
    show(img, adjust=True, title="Початковий об'єднаний файл (до репроекції)")
```

```python
# Репроекція у EPSG:4326
with rasterio.open(output_path) as src:
    transform, width, height = calculate_default_transform(
        src.crs, 'EPSG:4326', src.width, src.height, *src.bounds)
    profile = src.profile
    profile.update({
        'crs': 'EPSG:4326',
        'transform': transform,
        'width': width,
        'height': height
    })

    reprojected_path = r"D:\зпад\data7\sentinel_reprojected.tif"

    with rasterio.open(reprojected_path, 'w', **profile) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs='EPSG:4326',
                resampling=Resampling.nearest)

print("Репроекція завершена")
```

```python
# Показуємо зображення після репроекції
with rasterio.open(reprojected_path) as src:
    img = src.read([1,2,3])
    show(img, adjust=True, title="Після репроекції (без обрізання)")
```

```python
# Обрізання (імітація обрізки)
with rasterio.open(reprojected_path) as src:
    data = src.read()
    profile = src.profile

    margin = 500
    data_cropped = data[:, margin:-margin, margin:-margin]

    profile.update({
        'width': data_cropped.shape[2],
        'height': data_cropped.shape[1],
        'transform': src.transform * rasterio.Affine.translation(margin, margin)
    })

    cropped_path = r"D:\зпад\data7\sentinel_cropped.tif"
    with rasterio.open(cropped_path, 'w', **profile) as dst:
        dst.write(data_cropped)

print("Обрізання завершено")

```

```python

# Показуємо зображення після обрізання
with rasterio.open(cropped_path) as src:
    img = src.read([1,2,3])
    show(img, adjust=True, title="Після обрізання (красивий зріз)")
```
